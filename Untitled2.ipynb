```python
import os
# Ruta al directorio de imágenes
image_dir = os.path.join(dataset_path, 'data')

# Listar archivos en el directorio de imágenes
print("Archivos en el directorio de imágenes:")
print(os.listdir(image_dir))
```

```python
# Listar imágenes en cada subdirectorio
for subdir in os.listdir(image_dir):
    subdir_path = os.path.join(image_dir, subdir)
    if os.path.isdir(subdir_path):  # Asegurarse de que es un directorio
        print(f"Archivos en el subdirectorio '{subdir}':")
        print(os.listdir(subdir_path))
```

```python
# Contar imágenes en cada subdirectorio
for subdir in os.listdir(image_dir):
    subdir_path = os.path.join(image_dir, subdir)
    if os.path.isdir(subdir_path):
        image_files = [f for f in os.listdir(subdir_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        image_count = len(image_files)
        print(f"Subdirectorio '{subdir}' tiene {image_count} imágenes.")
```

```python
# Ruta al subdirectorio específico
specific_dir = os.path.join(image_dir, 'Straight')  #

import matplotlib.pyplot as plt
from PIL import Image

# Filtrar archivos de imagen en el subdirectorio específico
image_files = [f for f in os.listdir(specific_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
plt.figure(figsize=(20, 5))  # Cambia a un tamaño más ancho
num_images_to_show = 5
num_columns = 5
num_rows = (num_images_to_show + num_columns - 1) // num_columns  # Calcula el número de filas necesarias

plt.figure(figsize=(20, 10))  # Aumenta el tamaño de la figura para mayor claridad
for i, image_file in enumerate(image_files[:num_images_to_show]):
    img_path = os.path.join(specific_dir, image_file)
    try:
        img = Image.open(img_path)
        plt.subplot(num_rows, num_columns, i + 1)
        plt.imshow(img)
        plt.axis("off")  # Ocultar ejes
    except Exception as e:
        print(f"No se pudo abrir la imagen {image_file}: {e}")

plt.show()
```

VGG PARA LA CLASIFICACION DE IMAGENES

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator


# Parámetros de la carga de imágenes
img_height, img_width = 150, 150
batch_size = 32

# Generador de datos para cargar imágenes
datagen = ImageDataGenerator(rescale=1./255,
                             validation_split=0.2,  # Reservar un 20% para validación
                             rotation_range=20,
                             width_shift_range=0.2,
                             height_shift_range=0.2,
                             shear_range=0.2,
                             zoom_range=0.2,
                             horizontal_flip=True)

# Cargar datos de entrenamiento
train_generator = datagen.flow_from_directory(
    os.path.join(dataset_path, 'data'),
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='sparse',  # Cambiar a 'categorical' si las etiquetas son one-hot
    subset='training'      # Usar la porción de entrenamiento
)

# Cargar datos de validación
validation_generator = datagen.flow_from_directory(
    os.path.join(dataset_path, 'data'),
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='sparse',  # Cambiar a 'categorical' si las etiquetas son one-hot
    subset='validation'    # Usar la porción de validación
)
print("Clases en el generador de entrenamiento:", train_generator.class_indices)
print("Clases en el generador de validación:", validation_generator.class_indices)
```

```python
from tensorflow.keras import layers, models
from tensorflow.keras.applications import VGG16

# Cargar el modelo VGG16 preentrenado
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(img_height, img_width, 3))

# Congelar las capas base
for layer in base_model.layers:
    layer.trainable = False

# Añadir capas personalizadas
model = models.Sequential()
model.add(base_model)
model.add(layers.Flatten())
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dense(train_generator.num_classes, activation='softmax'))  # Número de clases en el dataset

# Compilar el modelo
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Entrenar el modelo
history = model.fit(train_generator,
                    epochs=10,
                    validation_data=validation_generator)
```

```python
# Gráfica de precisión
plt.plot(history.history['accuracy'], label='Precisión de Entrenamiento')
plt.plot(history.history['val_accuracy'], label='Precisión de Validación')
plt.title('Precisión del Modelo')
plt.ylabel('Precisión')
plt.xlabel('Epoch')
plt.legend(loc='upper left')
# Añadir valores numéricos a la gráfica de precisión
for i in range(len(history.history['accuracy'])):
    plt.text(i, history.history['accuracy'][i], f"{history.history['accuracy'][i]:.2f}", fontsize=8, ha='center', va='bottom')
    plt.text(i, history.history['val_accuracy'][i], f"{history.history['val_accuracy'][i]:.2f}", fontsize=8, ha='center', va='bottom')
plt.show()

# Gráfica de pérdida
plt.plot(history.history['loss'], label='Pérdida de Entrenamiento')
plt.plot(history.history['val_loss'], label='Pérdida de Validación')
plt.title('Pérdida del Modelo')
plt.ylabel('Pérdida')
plt.xlabel('Epoch')
plt.legend(loc='upper left')
for i in range(len(history.history['loss'])):
    plt.text(i, history.history['loss'][i], f"{history.history['loss'][i]:.2f}", fontsize=8, ha='center', va='bottom')
    plt.text(i, history.history['val_loss'][i], f"{history.history['val_loss'][i]:.2f}", fontsize=8, ha='center', va='bottom')
plt.show()
```
